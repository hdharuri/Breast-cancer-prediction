```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from time import time
# Import supplementary visualization code visuals.py
import visuals as vs
import xgboost as xgb
```

```python
# read in a csv file
data = pd.read_csv('dataR2.csv')
data.columns
data['Classification'] = data['Classification'].map({2:1,1:0})
data.groupby('Classification').size()
```

```python
# Split the data into features and target label
classification_raw = data['Classification']
features_raw = data.drop('Classification', axis = 1)
```

```python
# perform the mann whitney test to assess differences in 
# distribution of values for patients and controls  for each variable
from scipy.stats import mannwhitneyu

# subset the data for control and patient data
control_df = data[data.Classification == 0]
patient_df = data[data.Classification == 1]

# look at summary statistics based on groupby('Classification')
grouped_df = data.groupby('Classification').describe()

# This is to generate Table 1 of the project report
col_names =  ['Variable', 'Control', 'Patient','pvalue']
my_df  = pd.DataFrame(columns = col_names)

for var in grouped_df.columns.levels[0]:
    temp_df = grouped_df.get(var)    
    control, patient = temp_df[['mean']].values
    stat, p = mannwhitneyu(control_df[var], patient_df[var])
    my_dic = {'Variable':var, 'Control':control[0].round(2), 'Patient':patient[0].round(2),'pvalue':round(p,3)}
    my_df.loc[len(my_df)] = my_dic 

print(my_df)
# Generate Table 1 of the report
my_df.to_csv('descriptive_stats.txt',sep="\t",index=False)
```

```python
# Figure 2 of the report
features_raw.hist(figsize=(12,12))  # s is an instance of Series
plt.savefig('Histograms.jpg')
```

```python
# Log-transform the skewed features
skewed = ['Insulin','Adiponectin','Glucose','HOMA','Resistin','Leptin','MCP.1']
features_raw2 = data.drop('Classification', axis = 1)
features_log_transformed = data.drop('Classification', axis = 1)
features_log_transformed[skewed] = features_raw2[skewed].apply(lambda x: np.log(x + 1))

# Visualize the new log distributions
features_log_transformed.hist(figsize=(9,9))
```

```python
# Scaling the values
# Figure 3 of the report
from sklearn.preprocessing import MinMaxScaler

# Initialize a scaler, then apply it to the features
scaler = MinMaxScaler() # default=(0, 1)
columnNames = list(features_raw)

features_transformed = features_log_transformed
features_transformed[columnNames] = scaler.fit_transform(features_transformed[columnNames])
features_transformed.hist(figsize=(9,9))
plt.savefig('Histograms_transformed.jpg')
```

```python
# train-test split
features_final = features_transformed
# Import train_test_split
from sklearn.cross_validation import train_test_split

# Split the 'features' and 'income' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_final, 
                                                    classification_raw, 
                                                    test_size = 0.2, 
                                                    random_state = 95)

# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
# Import two metrics from sklearn - fbeta_score and accuracy_score
from sklearn.metrics import fbeta_score, accuracy_score
def train_predict(learner, X_train, y_train, X_test, y_test): 
    '''
    inputs:
       - learner: the learning algorithm to be trained and predicted on
       - sample_size: the size of samples (number) to be drawn from training set
       - X_train: features training set
       - y_train: income training set
       - X_test: features testing set
       - y_test: income testing set
    '''
    
    results = {}
    
    # Fit the learner to the training data using slicing with 'sample_size' using .fit(training_features[:], training_labels[:])
    start = time() # Get start time
    learner.fit(X_train,y_train)
    end = time() # Get end time
    
    # Calculate the training time
    results['train_time'] = end - start
        
    # Get the predictions on the test set(X_test),
    start = time() # Get start time
    predictions_test = learner.predict(X_test)
    predictions_train = learner.predict(X_train)
    end = time() # Get end time
    
    # Calculate the total prediction time
    results['pred_time'] = end - start
            
    # Compute accuracy on the training samples which is y_train
    results['acc_train'] = accuracy_score(y_train, predictions_train)
        
    # Compute accuracy on test set using accuracy_score()
    results['acc_test'] = accuracy_score(y_test,predictions_test)
    
    # Compute F-score on the the training samples using fbeta_score()
    results['f_train'] = fbeta_score(y_train,predictions_train,beta = 0.5)
        
    # Compute F-score on the test set which is y_test
    results['f_test'] = fbeta_score(y_test,predictions_test, beta = 0.5)
       
    # Success
    print("{} trained.".format(learner.__class__.__name__))
        
    # Return the results
    return results
```

```python
# TODO: Import the three supervised learning models from sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression

# TODO: Initialize the three models
clf_A = GaussianNB()
clf_B = AdaBoostClassifier(random_state=0)
clf_C = SVC(random_state=0)
clf_D = XGBClassifier()
clf_E = LogisticRegression()
clf_F = RandomForestClassifier()

# Collect results on the learners
results = {}
for clf in [clf_A, clf_B, clf_C, clf_D,clf_E, clf_F]:
    clf_name = clf.__class__.__name__
    results[clf_name] = {}
    results[clf_name] = \
    train_predict(clf, X_train, y_train, X_test, y_test)
    
results
```

```python
# Fine tune the model using GridSearchCV
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import make_scorer, fbeta_score, f1_score, auc,roc_curve
# Explore XGBoost, fine tune parameters etc
classifier = XGBClassifier()
parameters = [{'learning_rate':[0.1,0.2,0.3,0.5], 'max_depth':[1,2,3,4,5,6,7]}]

scorer = make_scorer(fbeta_score,beta=0.5)

grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = scorer,
                           cv = 5,
                           n_jobs = -1
                          )
grid_search = grid_search.fit(X_train, y_train)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
best_classifier = grid_search.best_estimator_

print(best_parameters)
print(best_classifier)
```

```python
# Predict the test labels using the optimized classifier
best_predict = best_classifier.predict(X_test)
#y_test_arr = y_test.values
#y_test_arr = np.concatenate((1-best_predict.reshape(-1,1),best_predict.reshape(-1,1)),axis=1)

fbetaScore = fbeta_score(y_test, best_predict, beta = 0.5)
print(fbetaScore)

# TABLE 2
# Generate the Confusion Matrix
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, best_predict).ravel()
(tn, fp, fn, tp)
```

```python
# FIGURE 6
# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
fpr, tpr, _ = roc_curve(y_test, best_predict)
roc_auc = auc(fpr, tpr)

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Roc_curve.jpg')
```

```python
# FIGURE 5
# Import a supervised learning model that has 'feature_importances_'
# Train the supervised model on the training set using .fit(X_train, y_train)
model = XGBClassifier(random_state=9).fit(X_train,y_train)

# Extract the feature importances using .feature_importances_ 
importances = model.feature_importances_
importances
# Plot
vs.feature_plot(importances, X_train, y_train)
fig.savefig('feature_importance.jpg')
```

```python
# Make predictions on the reduced dataset with only the important features determined above
# Import functionality for cloning a model
from sklearn.base import clone

# Reduce the feature space
X_train_reduced = X_train[X_train.columns.values[(np.argsort(importances)[::-1])[:5]]]
X_test_reduced = X_test[X_test.columns.values[(np.argsort(importances)[::-1])[:5]]]

# Train on the "best" model found from grid search earlier
clf = (clone(best_classifier)).fit(X_train_reduced, y_train)

# Make new predictions
reduced_predictions = clf.predict(X_test_reduced)

print("\nFinal Model trained on reduced data\n------")
print("Accuracy on testing data: {:.4f}".format(accuracy_score(y_test, reduced_predictions)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, reduced_predictions, beta = 0.5)))
```

```python
# Ascertain that raw data for the reduced dataset performs identically to the above with transformed values
xtrain_ind = X_train.index.values
X_train_raw = features_raw.iloc[xtrain_ind]

xtest_ind = X_test.index.values
X_test_raw = features_raw.iloc[xtest_ind]


# Reduce the feature space
X_train_reduced_raw = X_train_raw[X_train_raw.columns.values[(np.argsort(importances)[::-1])[:5]]]
X_test_reduced_raw = X_test_raw[X_test_raw.columns.values[(np.argsort(importances)[::-1])[:5]]]

# Train on the "best" model found from grid search earlier
clf_raw = (clone(best_classifier)).fit(X_train_reduced_raw, y_train)

# Make new predictions
reduced_predictions = clf_raw.predict(X_test_reduced_raw)

print("\nFinal Model trained on reduced data\n------")
print("Accuracy on testing data: {:.4f}".format(accuracy_score(y_test, reduced_predictions)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, reduced_predictions, beta = 0.5)))
```
